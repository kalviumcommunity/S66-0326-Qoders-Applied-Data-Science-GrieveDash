# Code vs Markdown Demonstration

**Purpose**: This notebook demonstrates the intentional use of both Code and Markdown cells. It showcases how to separate executable logic (code) from human-readable explanations (markdown).

In a professional Data Science workflow:
- **Markdown cells** are used to provide narrative, reasoning, and document our thought processes.
- **Code cells** are used purely for executing the Python logic required for the analysis.

### 1. Simple Arithmetic Operation

```python
a = 15
b = 27
total = a + b
print(f"The sum of {a} and {b} is: {total}")
```

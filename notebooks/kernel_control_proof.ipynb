# Kernel Control Proof

This notebook demonstrates how to manage the Jupyter Kernel intentionally. We will show:
1. **Execution Order** and how variables retain state based on the order cells are executed.
2. **Restarting the Kernel** to clear state and maintain predictability.
3. **Interrupting Execution** of a long-running process.

--- 
### 1. Execution Order & Hidden State
First, we define a variable `counter` and assign it a value of `0`.

```python
counter = 0
print(f"Initial counter value: {counter}")
```

Next, we increment the variable. If you run this cell multiple times without running the cell above, the variable will keep incrementing. This demonstrates that the kernel retains state sequentially.

```python
counter += 5
print(f"Current counter value: {counter}")
```

### 2. Restarting the Kernel

If our variable `counter` has grown unexpectedly, we can reset it by restarting the kernel using the Jupyter menu (`Kernel -> Restart Kernel`). After restarting, all variables are erased from memory and we must run the notebook from the top to ensure clean, predictable data.

Try it: Restart the kernel and run the cell below without running the first cell. It will throw a `NameError` because the state was wiped out!

```python
try:
    print(f"After restart, counter is: {counter}")
except NameError as e:
    print(f"Expected Error: {e} - The kernel state was successfully wiped!")
```

### 3. Interrupting a Long-Running Process

Sometimes, a cell takes far too long to compute or gets stuck in a loop. In these cases, we must use `Kernel -> Interrupt Kernel` (the stop button).

Below is a deliberately slow counting loop. It will take 10 seconds to finish on its own, but you can manually interrupt it while it is running.

```python
import time

print("Starting a deliberately long-running process...")
try:
    for i in range(1, 11):
        print(f"Processing step {i}/10...")
        time.sleep(1) # Wait for 1 second per step
    print("Process completed successfully.")
except KeyboardInterrupt:
    print("\n--- Execution Safely Interrupted! ---")
    print("The notebook remains responsive and did not freeze.")
```

# Python Numeric and String Data Types

**Milestone**: Understanding Python Data Types

This notebook explores Python's foundational data types — integers, floats, and strings — which underpin all data analysis work. Understanding these correctly prevents logical errors, type mismatches, and unexpected program behaviour.

---
## 1. Integer Data Type (`int`)

Integers represent whole numbers. They are used for counts, IDs, and quantities that cannot have decimals.

```python
total_complaints = 240
resolved_complaints = 185
pending_complaints = total_complaints - resolved_complaints

print(f"Type of total_complaints : {type(total_complaints)}")
print(f"Total    : {total_complaints}")
print(f"Resolved : {resolved_complaints}")
print(f"Pending  : {pending_complaints}")
```

---
## 2. Float Data Type (`float`)

Floats represent decimal numbers. They are used for rates, percentages, averages, and any measurement requiring precision.

```python
resolution_rate = resolved_complaints / total_complaints * 100
avg_days_to_resolve = 4.75

print(f"Type of resolution_rate    : {type(resolution_rate)}")
print(f"Resolution Rate            : {round(resolution_rate, 2)}%")
print(f"Avg Days to Resolve        : {avg_days_to_resolve} days")

# Integer division vs true division
print(f"\n10 // 3 (floor division)   : {10 // 3}   → result is int")
print(f"10  / 3 (true division)    : {10  / 3:.4f} → result is float")
```

---
## 3. String Data Type (`str`)

Strings represent text. They are essential for category labels, complaint descriptions, status fields, and any human-readable data.

```python
project_name  = "GrieveDash"
complaint_cat = "Water Supply"
status        = "open"

print(f"Type of project_name : {type(project_name)}")

# Concatenation
label = complaint_cat + " — " + status.upper()
print(f"Concatenated label   : {label}")

# f-string formatting
summary = f"[{project_name}] Category: {complaint_cat} | Status: {status.capitalize()}"
print(f"Formatted summary    : {summary}")

# String methods
print(f"Length of label      : {len(label)} characters")
print(f"Uppercase label      : {label.upper()}")
```

---
## 4. Type Mismatch — Error and Conversion

Python does **not** allow mixing strings and numbers directly. Below we observe the error, then fix it with explicit type conversion using `str()` and `int()`.

```python
# Demonstrate type mismatch
try:
    bad_result = "Total complaints: " + total_complaints   # int + str → TypeError
except TypeError as e:
    print(f"TypeError caught: {e}")

# Fix: convert number to string
good_result = "Total complaints: " + str(total_complaints)
print(f"Fixed with str()  : {good_result}")

# Fix: convert string to number
count_from_input = "47"                     # simulates input() or CSV read
count_as_int     = int(count_from_input)
print(f"String '47' + 10  : {count_as_int + 10}  (after int() conversion)")
```

---
## 5. Inspecting Data Types at Runtime

Always verify the type of a variable when something behaves unexpectedly. `type()` is your first debugging tool.

```python
samples = [
    ("total_complaints",   total_complaints),
    ("resolution_rate",    resolution_rate),
    ("project_name",       project_name),
    ("avg_days_to_resolve",avg_days_to_resolve),
    ("count_from_input",   count_from_input),
    ("count_as_int",       count_as_int),
]

print(f"{'Variable':<25} {'Value':<15} {'Type'}")
print("-" * 55)
for name, val in samples:
    print(f"{name:<25} {str(val):<15} {type(val).__name__}")
```

## ✅ Summary

| Data Type | Python Keyword | Example | Use Case |
|-----------|---------------|---------|----------|
| Integer   | `int`         | `240`   | Counts, IDs |
| Float     | `float`       | `4.75`  | Rates, averages |
| String    | `str`         | `"open"` | Labels, text fields |

**Key Rules:**
- Use `type()` to inspect any variable.
- Use `str()`, `int()`, `float()` for explicit type conversion.
- Never assume a variable's type — verify it, especially when data comes from CSV files or user input.

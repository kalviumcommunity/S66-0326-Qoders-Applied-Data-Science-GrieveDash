# Python Lists, Tuples, and Dictionaries

**Milestone**: Working with Python Collections  
**Project**: GrieveDash — Applied Data Science Sprint

This notebook explores Python's three core collection types:
- **Lists** — ordered, mutable sequences
- **Tuples** — ordered, immutable sequences  
- **Dictionaries** — key-value mappings

Choosing the right structure for the right data prevents bugs and makes code more readable.

---
## 1. Lists — Ordered & Mutable

Lists store multiple items in a specific order and can be changed after creation. Use them for dynamic collections where items are added, removed, or updated.

### 1.1 Creating and Accessing

```python
# A list of open grievance complaint IDs
open_complaints = [1021, 1034, 1045, 1056, 1078]

print(f"Type         : {type(open_complaints)}")
print(f"All IDs      : {open_complaints}")
print(f"First ID     : {open_complaints[0]}   (index 0)")
print(f"Last ID      : {open_complaints[-1]}  (index -1)")
print(f"Slice [1:3]  : {open_complaints[1:3]}")
```

### 1.2 Modifying a List

Lists are **mutable** — we can append, update, and remove elements at any time.

```python
# Add a new complaint
open_complaints.append(1099)
print(f"After append : {open_complaints}")

# Resolve complaint 1034 — remove it
open_complaints.remove(1034)
print(f"After remove : {open_complaints}")

# Update an existing ID
open_complaints[0] = 1022
print(f"After update : {open_complaints}")

print(f"Total open   : {len(open_complaints)} complaints")
```

---
## 2. Tuples — Ordered & Immutable

Tuples store fixed collections of values that should not change. Use them for constants, configuration settings, and data integrity-critical values.

```python
# Fixed complaint categories — should never change at runtime
COMPLAINT_CATEGORIES = ("water", "roads", "electricity", "sanitation", "noise")

print(f"Type            : {type(COMPLAINT_CATEGORIES)}")
print(f"All categories  : {COMPLAINT_CATEGORIES}")
print(f"First category  : {COMPLAINT_CATEGORIES[0]}")
print(f"Total categories: {len(COMPLAINT_CATEGORIES)}")
```

### 2.1 Demonstrating Immutability

Attempting to modify a tuple raises a `TypeError`, which protects our data from accidental changes.

```python
try:
    COMPLAINT_CATEGORIES[0] = "flooding"   # Not allowed!
except TypeError as e:
    print(f"TypeError caught: {e}")
    print("✅ Tuple immutability correctly prevented accidental modification.")
```

---
## 3. Dictionaries — Key-Value Mappings

Dictionaries store data as `key: value` pairs. They are ideal for representing structured records — like a single complaint entry or a summary report.

```python
# A single grievance record as a dictionary
complaint = {
    "id"      : 1021,
    "category": "water",
    "status"  : "open",
    "priority": "high",
    "ward"    : "Ward 7"
}

print(f"Type           : {type(complaint)}")
print(f"Complaint ID   : {complaint['id']}")
print(f"Category       : {complaint['category'].capitalize()}")
print(f"Status         : {complaint['status'].upper()}")
```

### 3.1 Modifying a Dictionary

Dictionaries are mutable — we can update existing keys or add new ones.

```python
# Update existing key
complaint["status"] = "resolved"

# Add new key
complaint["resolved_by"] = "Field Team A"

print("Updated complaint record:")
for key, value in complaint.items():
    print(f"   {key:<15}: {value}")
```

## 4. Choosing the Right Data Structure

The table below summarises when to use each collection type:

```python
comparison = [
    ("List",       "Ordered", "Mutable",   "Index", "Dynamic collections — complaint IDs"),
    ("Tuple",      "Ordered", "Immutable", "Index", "Fixed configs — valid categories"),
    ("Dictionary", "Unordered","Mutable",  "Key",   "Structured records — complaint entry"),
]

header = f"{'Type':<12} {'Order':<10} {'Mutable':<12} {'Access':<8} Use Case"
print(header)
print("-" * 70)
for row in comparison:
    print(f"{row[0]:<12} {row[1]:<10} {row[2]:<12} {row[3]:<8} {row[4]}")
```

---
## ✅ Summary

- **Lists** are the go-to for dynamic ordered data. Use `append()`, `remove()`, and indexing freely.
- **Tuples** are for fixed data that must not change — like valid category names or app constants. Immutability is a safety feature, not a limitation.
- **Dictionaries** model structured entities with named fields. Use them whenever data has labels.

> **Golden Rule**: If the data should be read-only → use a **tuple**. If it needs labels → use a **dict**. If it's a dynamic sequence → use a **list**.

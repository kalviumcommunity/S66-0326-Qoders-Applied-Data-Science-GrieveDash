## 5. Quick Demo — NumPy & Pandas

```python
# NumPy array operations
arr = np.array([10, 20, 30, 40, 50])
print("NumPy array :", arr)
print("Mean        :", arr.mean())
print("Std Dev     :", round(arr.std(), 2))
```

```python
# Pandas DataFrame demo
data = {
    "Category": ["Roads", "Water", "Electricity", "Sanitation", "Noise"],
    "Complaints": [120, 85, 64, 97, 42],
    "Resolved": [98, 70, 55, 80, 38],
}
df = pd.DataFrame(data)
df["Resolution %"] = round(df["Resolved"] / df["Complaints"] * 100, 1)

print("Sample Grievance Data:")
df
```

---
## 6. Quick Demo — Matplotlib & Seaborn Plot

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8, 4))

x = np.arange(len(df))
width = 0.35

bars1 = ax.bar(x - width/2, df["Complaints"], width, label="Complaints", color="#e74c3c")
bars2 = ax.bar(x + width/2, df["Resolved"],   width, label="Resolved",   color="#2ecc71")

ax.set_xlabel("Category")
ax.set_ylabel("Count")
ax.set_title("GrieveDash — Sample Complaints vs Resolved")
ax.set_xticks(x)
ax.set_xticklabels(df["Category"])
ax.legend()
ax.grid(axis="y", alpha=0.3)

plt.tight_layout()
plt.show()
```
